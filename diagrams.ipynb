```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Relu Activation Function DFiagram

# Define the ReLU function
def relu(x):
    return np.maximum(0, x)

# Generate x values
x = np.linspace(-10, 10, 100)

# Apply ReLU function to x
y = relu(x)

# Create the plot
plt.plot(x, y, label='ReLU')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.title('ReLU Activation Function')
plt.legend()

# Show the plot
plt.show()






```

```python
# Linear Activation Function Diagram

# Define the Linear activation function
def linear(x):
    return x

# Generate x values
x = np.linspace(-10, 10, 100)

# Apply Linear function to x
y = linear(x)

# Create the plot
plt.plot(x, y, label='Linear')
plt.xlabel('x')
plt.ylabel('Linear(x)')
plt.title('Linear Activation Function')
plt.legend()

# Show the plot
plt.show()
```

```python
# Define the Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate x values
x = np.linspace(-10, 10, 100)

# Apply Sigmoid function to x
y = sigmoid(x)

# Create the plot
plt.plot(x, y, label='Sigmoid')
plt.xlabel('x')
plt.ylabel('Sigmoid(x)')
plt.title('Sigmoid Activation Function')
plt.legend()

# Show the plot
plt.show()
```

```python
# Define the SoftPlus activation function
def softplus(x):
    return np.log(1 + np.exp(x))

# Generate x values
x = np.linspace(-10, 10, 100)

# Apply SoftPlus function to x
y = softplus(x)

# Create the plot
plt.plot(x, y, label='SoftPlus')
plt.xlabel('x')
plt.ylabel('SoftPlus(x)')
plt.title('SoftPlus Activation Function')
plt.legend()

# Show the plot
plt.show()
```

```python
# Define the Tanh activation function
def tanh(x):
    return np.tanh(x)

# Generate x values
x = np.linspace(-10, 10, 100)

# Apply Tanh function to x
y = tanh(x)

# Create the plot
plt.plot(x, y, label='Tanh')
plt.xlabel('x')
plt.ylabel('Tanh(x)')
plt.title('Tanh Activation Function')
plt.legend()

# Show the plot
plt.show()
```

```python
# Generate random data
np.random.seed(0)
X = np.random.rand(100) * 10
y = 2 * X + np.random.randn(100)

# Perform linear regression
coefficients = np.polyfit(X, y, 1)
m, b = coefficients

# Generate predicted values
y_pred = m * X + b

# Create the plot
plt.scatter(X, y, label='Data')
plt.plot(X, y_pred, color='red', label='Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()

# Show the plot
plt.show()
```
