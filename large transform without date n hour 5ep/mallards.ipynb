```python
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn import tree

mallards=pd.read_excel("ConstrainedSet_.xlsx")
```

```python
print(mallards.head(5))
```

```python
df=pd.DataFrame(mallards)
features=df[["population", "farmed", "date", "hour", "mid", "slatitude", "slongitude", "v2", "v1"]]
print(features)
```

```python
ordered = features.sort_values(by=["population"])
ordered
```

```python
data=[]
# preddata = ["population", "farmed", "date", "hour", "mid", "slatitude", "slongitude", "v2"]
preddata = pd.DataFrame(columns=["population", "farmed", "slatitude", "slongitude", "v2", "dlatitude", "dlongitude"])

for x in range(0,len(ordered)):
    if(x==0):
        preddata.loc[x, "slongitude"] = ordered.loc[x,"slongitude"]-13
        preddata.loc[x, "dlongitude"] = ordered.loc[x,"slongitude"]-13
        preddata.loc[x, "slatitude"] = ordered.loc[x,"slatitude"]-56
        preddata.loc[x, "dlatitude"] = ordered.loc[x,"slatitude"]-56
        preddata.loc[x, "population"] = ordered.loc[x,"population"]/1000
        preddata.loc[x, "farmed"] = ordered.loc[x,"farmed"]
        preddata.loc[x, "v2"] = ordered.loc[x,"v2"]/1000
        print( preddata.loc[x])
    elif(x==len(ordered)-1):
        break;
    #preddata.append([features.loc[:x,"slatitude"], features.loc[:x,"slongitude"]])
    elif(ordered.loc[x, "population"] != ordered.loc[x+1, "population"]):
        preddata.loc[x, "slongitude"] = ordered.loc[x,"slongitude"]-13
        preddata.loc[x, "dlongitude"] = ordered.loc[x,"slongitude"]-13
        preddata.loc[x, "slatitude"] = ordered.loc[x,"slatitude"]-56
        preddata.loc[x, "dlatitude"] = ordered.loc[x,"slatitude"]-56
        preddata.loc[x, "population"] = ordered.loc[x,"population"]/1000
        preddata.loc[x, "farmed"] = ordered.loc[x,"farmed"]
        preddata.loc[x, "v2"] = ordered.loc[x,"v2"]/1000
        print( preddata.loc[x])
    else:
        preddata.loc[x, "slongitude"] = ordered.loc[x,"slongitude"]-13
        preddata.loc[x, "dlongitude"] = ordered.loc[x+1,"slongitude"]-13
        preddata.loc[x, "slatitude"] = ordered.loc[x,"slatitude"]-56
        preddata.loc[x, "dlatitude"] = ordered.loc[x+1,"slatitude"]-56
        preddata.loc[x, "population"] = ordered.loc[x,"population"]/1000
        preddata.loc[x, "farmed"] = ordered.loc[x,"farmed"]
        preddata.loc[x, "v2"] = ordered.loc[x,"v2"]/1000
        print(preddata.loc[x,"slongitude"])

print("=================================")
```

```python
X=preddata[["population", "farmed", "slatitude", "slongitude", "v2"]]
y=preddata[["dlatitude"]]
print(X.head(5))
print(y.head(5))
```

```python
#Importing the libraries
import tensorflow as tf
from tensorflow import keras
from keras.layers import Dense

tf.__version__
keras.__version__

#tf.keras.layers.experimental.preprocessing.Normalization
# normalizer = tf.keras.layers.Normalization(axis=-1)
normalizer = tf.keras.layers.experimental.preprocessing.Normalization(axis=-1)
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(10000, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(1000, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(100, activation=tf.nn.softmax))

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
y = y.astype(np.float32)
X = X.astype(np.float32)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train ,y_test=train_test_split(X, y, test_size=0.20)
```

```python
model.layers
```

```python
history=model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=5)
```
